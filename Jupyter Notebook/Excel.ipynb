```python
# 帮助 白白 删除成绩表中有缺考标记的行
import win32com.client   #这里用到win32com.client，需要安装pywin32模块
                         #下载地址 http://sourceforge.net/projects/pywin32/files/pywin32/Build216/
xlApp = win32com.client.Dispatch('Excel.Application')  #打开EXCEL，这里不需改动
xlBook = xlApp.Workbooks.Open('C:\\Users\\lenovo\\Desktop\\grade.xlsx')             #将D:\\1.xls改为要处理的excel文件路径
xlSht = xlBook.Worksheets('Sheet1')                    #要处理的excel页，默认第一页是‘Sheet1’
n=317      #n改为要处理的表格内容的行数
m=15       #m改为要处理的表格的内容的列数
try:      #为了加强程序健壮性，使用try...except方式避免出错后程序中断。可以将try、except删去。
     i=2
     while i<=n:
        temp3 = str(xlSht.Cells(i,3).Value)      #取每行的第3列
        temp5 = str(xlSht.Cells(i,5).Value)       #取每行的第5列
        temp7 = str(xlSht.Cells(i,7).Value)       #取每行的第7列
        temp9 = str(xlSht.Cells(i,9).Value)      #取每行的第9列
        
        if (temp3.find('缺考') > -1 or temp5.find('缺考') > -1 or temp7.find('缺考') > -1 or temp9.find('缺考') > -1 ):  #表示找到了字符串'缺考'
            print('被删除的行' + str(i))
            print()
            xlSht.Rows(i).Delete()        #将满足条件的元素所在行 整行删除
            n=n-1     #由于删除了一行，同时修改相关参数，避免判断遗漏
            continue
        
        print('没被删除的行' + str(i))
        print()
        i=i+1     #行标加1，继续判断下一行
except:
     print ('runerror')
xlBook.Close(SaveChanges=1)   #完成 关闭保存文件
del xlApp
```
